```python
df1 = pd.read_csv('ÏÑ±Ï†ÅÏùÑ Î∂ÄÌÉÅÌï¥ Ìã∞Ï≥êÏä§.csv')
columns_to_convert = ['view_count', 'like_count']
df1[columns_to_convert] = df1[columns_to_convert].apply(pd.to_numeric, errors='coerce')

channels_of_interest = ['play Ï±ÑÎÑêA', 'Ï±ÑÎÑêÏïÑÌïò: Ï±ÑÎÑêA Health & Asset', 'Ï±ÑÎÑêÏóêÏù¥Îìú: Ï±ÑÎÑêA Drama & Enjoy', 'Ï±ÑÎÑêA WORLD']
df1_filtered = df1[df1['channel_name'].isin(channels_of_interest)]
df1_filtered = df1_filtered.reset_index(drop=True)

# 'published' Ïó¥ÏùÑ datetime ÌòïÏãùÏúºÎ°ú Î≥ÄÌôò
df1_filtered['published'] = pd.to_datetime(df1_filtered['published'])

# ÏõîÎ≥ÑÎ°ú Í∑∏Î£πÌôî
df1_filtered['month'] = df1_filtered['published'].dt.to_period('M')

# ÏõîÎ≥Ñ ÌÜµÍ≥ÑÎüâ Í≥ÑÏÇ∞
month_stats = df1_filtered.groupby('month')['view_count'].agg([
    'min',            # ÏµúÏÜåÍ∞í
    lambda x: x.quantile(0.25),  # 1Î∂ÑÏúÑÍ∞í
    'median',         # Ï§ëÏïôÍ∞í
    lambda x: x.quantile(0.75),  # 3Î∂ÑÏúÑÍ∞í
    'max'             # ÏµúÎåÄÍ∞í
]).reset_index()

# Ïó¥ Ïù¥Î¶Ñ Î≥ÄÍ≤Ω
month_stats.columns = ['Month', 'Min', '1st Quartile', 'Median', '3rd Quartile', 'Max']

# Í≤∞Í≥º Ï∂úÎ†•
print(month_stats)
```

```python
plt.figure(figsize=(14, 8))

# ÎßâÎåÄ Í∑∏ÎûòÌîÑ
plt.subplot(2, 1, 1)
month_stats.plot(x='Month', kind='bar', stacked=True, colormap='viridis', ax=plt.gca())
plt.title('Monthly View Count Statistics (Bar Plot)')
plt.xlabel('Month')
plt.ylabel('View Count')
plt.legend(title='Statistics')

# ÏÑ† Í∑∏ÎûòÌîÑ
plt.subplot(2, 1, 2)
plt.plot(month_stats['Month'].astype(str), month_stats['Min'], marker='o', label='Min')
plt.plot(month_stats['Month'].astype(str), month_stats['1st Quartile'], marker='o', label='1st Quartile')
plt.plot(month_stats['Month'].astype(str), month_stats['Median'], marker='o', label='Median')
plt.plot(month_stats['Month'].astype(str), month_stats['3rd Quartile'], marker='o', label='3rd Quartile')
plt.plot(month_stats['Month'].astype(str), month_stats['Max'], marker='o', label='Max')
plt.title('Monthly View Count Statistics (Line Plot)')
plt.xlabel('Month')
plt.ylabel('View Count')
plt.legend()
plt.xticks(rotation=45)

plt.tight_layout()

# plt.savefig('Ìã∞Ï≤òÏä§_ÏõîÎ≥Ñ Ï°∞ÌöåÏàò ÏöîÏïΩ')

plt.show()
```
